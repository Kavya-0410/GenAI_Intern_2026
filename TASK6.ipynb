Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking_system(vehicles_logs, capacity):
    current_count=0     # How many vehicles are inside now
    peak_count=0        # highest vehicles at any time

    for log in vehicles_logs:
        if log =="IN":
            current_count+=1
        elif log == "OUT":
            if current_count>0:
                current_count-=1

        if current_count>peak_count:
            peak_count=current_count
            
    print("Currently Parked Vehicles:", current_count)
    
    if current_count>capacity:
        print("parking status:exceeded capacity")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_count)   

    
```

```python
smart_parking_system(["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"], 50)
                   
```

```python
def delivery_time(distance, weather, traffic):
    delivery_time=distance*5        #Distance (km): 8.....8*5=40

    if traffic =="High":
        delivery_time+=15    #40+15=55
    elif traffic =="Medium":
        delivery_time+=10
    
    print("Estimated Delivery Time:", delivery_time, "minutes")
    
```

 Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy(total_seats, booked_seats):

    booked_count=len(booked_seats)    #len(booked_seats) → gives number of booked seats
    occupancy=(booked_count/total_seats)*100

    print("Occupancy:",round(occupancy), "%")
    #Python uses banker's rounding:
    #74.5 → 74
    #75.5 → 76
    #It rounds to the nearest even number when exactly .5

    if occupancy==100:
        print("show status:Houseful")
    elif occupancy>=75:
        print("show status:Almost full")
        print("suggestion:consider opening additional show")
    else:
        print("show status:available")
```

```python
seat_occupancy(200, [1]*150)
```

Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    average_cpu = sum(cpu_readings) / len(cpu_readings)
    print("Average CPU Load:", round(average_cpu), "%")

    if average_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
```

```python
classify_server_load([45, 60, 70, 85, 90])
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
```

```python
classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

Problem 6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):


    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
```

```python
event_registration_controller(100, 105)
```
